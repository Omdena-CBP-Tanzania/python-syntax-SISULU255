```python
students = {"Egito": 73,
            "Hussein": 67,
            "Majobo": 82,
            "WALTER":99}

students_above_80 = []

for student in students:
    if students[student] > 80:
        students_above_80.append(student)

# Print the result
print(students_above_80)
```
